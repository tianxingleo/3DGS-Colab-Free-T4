```python
!mkdir -p colmap_output

# 特征提取（禁用 GPU，禁用 GUI）
!colmap feature_extractor \
    --SiftExtraction.use_gpu=0 \
    --database_path colmap_output/database.db \
    --image_path frames

# 匹配（禁用 GPU）
!colmap exhaustive_matcher \
    --SiftMatching.use_gpu=0 \
    --database_path colmap_output/database.db

# SfM 重建
!mkdir -p colmap_output/sparse

!colmap mapper \
    --database_path colmap_output/database.db \
    --image_path frames \
    --output_path colmap_output/sparse \
    --Mapper.num_threads=8
```

```python
import os
print(os.listdir("colmap_output/sparse"))
```

```python
!mkdir -p colmap_output/sparse_text

!colmap model_converter \
    --input_path colmap_output/sparse/0 \
    --output_path colmap_output/sparse_text \
    --output_type TXT
!ns-train splatfacto \
    --data frames \
    --pipeline.model.eval-num-rays-per-chunk 8192 \
    --pipeline.model.camera-optimizer.mode off \
    --viewer.quit-on-train-completion True
```

```python
!ns-train splatfacto colmap \
    --data frames \
    --colmap-path colmap_output/sparse_text
```
